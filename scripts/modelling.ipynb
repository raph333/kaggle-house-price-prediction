# trying simple linear regression with (un-scaled) numeric features

```python
numeric_features = [x for x in train.columns if np.issubdtype(train[x], np.number) and x not in ('Id', 'SalePrice') ]
numeric_ds = train[[*numeric_features, 'Id', 'SalePrice']]
numeric_ds.index = numeric_ds.Id

numeric_ds[numeric_features] = numeric_ds[numeric_features].apply(preprocessing.scale_feature, axis=0)
numeric_ds = preprocessing.impute_GarageYrBlt(numeric_ds)
numeric_ds = preprocessing.impute_MasVnrArea(numeric_ds)
numeric_ds = preprocessing.impute_LotFrontage(numeric_ds)
numeric_ds = numeric_ds.dropna()

numeric_ds_train = numeric_ds.sample(frac=0.75)
numeric_ds_val = numeric_ds[~numeric_ds.index.isin(numeric_ds_train.index)]
X = np.array(numeric_ds_train[numeric_features])
y = numeric_ds_train[['SalePrice']]
lm = linear_model.LinearRegression()
lm.fit(X, y)

X_val = np.array(numeric_ds_val[numeric_features])
y_val = numeric_ds_val['SalePrice']
prediction = lm.predict(X_val)
prediction = [x for sub in prediction for x in sub]
evaluation_df = pd.DataFrame(y_val).assign(**{'prediction': prediction})
evaluation_df.prediction = evaluation_df.prediction.map(round).astype(int)
evaluation_df['diff'] = (evaluation_df.SalePrice - evaluation_df.prediction)
evaluation_df['sqrt_error'] = (evaluation_df.SalePrice - evaluation_df.prediction)**2
MSE = sum(evaluation_df.sqrt_error) / len(evaluation_df)
print(MSE)
```

```python
from sklearn.model_selection import cross_val_predict
X = np.array(numeric_ds[numeric_features])
y = numeric_ds['SalePrice']

pred = cross_val_predict(linear_model.LinearRegression(), X, y)
eval_df = pd.DataFrame(y).assign(**{'pred': [int(round(x)) for x in pred]})
eval_df['sqrt_error'] = (eval_df.SalePrice - eval_df.pred)**2
mse = sum(eval_df.sqrt_error) / len(eval_df)
print('MSE: {:.0f}*10⁶'.format(mse / 1000000))
```
